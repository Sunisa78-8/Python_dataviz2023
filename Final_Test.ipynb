#Merge Table

```python
import pandas as pd

# รวม (merge) ตารางโดยใช้คอลัมน์ที่เกี่ยวข้องแบบต่อกันทางด้านขวา
data = pd.merge(modified_transactions, operator, on='OperatorID', how='right')
```

```python
import os

# ตรวจสอบไดเรกทอรีปัจจุบัน
current_directory = os.getcwd()

# พิมพ์ที่อยู่ไดเรกทอรีปัจจุบัน
print("Current Directory:", current_directory)

# ค้นหาไฟล์ Excel
target_file = "modified_transactions.xlsx"  # แทนที่ด้วยชื่อไฟล์ที่คุณต้องการค้นหา
for root, dirs, files in os.walk(current_directory):
    if target_file in files:
        file_path = os.path.join(root, target_file)
        print("File found at:", file_path)
        break
```
